```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")

# Convert OrderDate column to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# --- 1. Revenue by City ---
city_sales = df.groupby('City')['TotalAmount'].sum().sort_values(ascending=False)
print("Revenue by City:\n", city_sales)

plt.figure(figsize=(8,5))
city_sales.plot(kind='bar', title="Revenue by City")
plt.ylabel("Revenue")
plt.show()

# --- 2. Top 10 Products by Revenue ---
top_products = df.groupby('Product')['TotalAmount'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products by Revenue:\n", top_products)

plt.figure(figsize=(8,5))
top_products.plot(kind='barh', title="Top 10 Products by Revenue")
plt.xlabel("Revenue")
plt.show()

# --- 3. Monthly Sales Trend ---
monthly_sales = df.groupby(df['OrderDate'].dt.to_period("M"))['TotalAmount'].sum()
print("\nMonthly Sales Trend:\n", monthly_sales)

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line', marker='o', title="Monthly Sales Trend")
plt.ylabel("Revenue")
plt.show()
```
